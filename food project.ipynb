```python


# Calculate average food preparation time for each restaurant
avg_prep_time = data.groupby('restaurant_name')['food_preparation_time'].mean()
print("Average food preparation time for each restaurant:")
print(avg_prep_time)

# Calculate average delivery time for each restaurant
avg_delivery_time = data.groupby('restaurant_name')['delivery_time'].mean()
print("\nAverage delivery time for each restaurant:")
print(avg_delivery_time)

# Observations
print("\nObservations:")
print("1. Restaurant Alidoro has the highest average food preparation time.")
print("2. Restaurant Zero Otto Nove has the lowest average delivery time.")
```

```python

# Drop rows with 'Not given' in the 'rating' column
df = data[data['rating'] != 'Not given']

# Convert the 'rating' column to numeric
data['rating'] = pd.to_numeric(data['rating'], errors='coerce')

# Calculate average rating of each restaurant
avg_rating = data.groupby('restaurant_name')['rating'].mean()
print("Average rating of each restaurant:")
print(avg_rating)

# Calculate correlation between food preparation time, delivery time, and customer ratings
correlation_matrix = data[['food_preparation_time', 'delivery_time', 'rating']].corr()
print("\nCorrelation Matrix:")
print(correlation_matrix)
```
